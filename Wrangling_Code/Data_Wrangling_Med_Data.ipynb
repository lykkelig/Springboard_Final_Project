```python
# Function to apply replacements with modification to handle names with more than one space
def replace_first_name(name):
    parts = name.split(' ', maxsplit=1)  # Split at the first space only
    if len(parts) == 2:  # Ensure there are two parts
        first, rest = parts
        first = replacements.get(first, first)  # Replace the first name if in replacements
        return f"{first} {rest}"
    else:
        return name  # Return the original name if it doesn't split into two parts
```

```python
# Apply the function to the 'name' column
rcd_df['name'] = rcd_df['name'].apply(replace_first_name)

print(rcd_df)
```

```python
# Create realistic medical data for medicines but we'll just use supplements for now
import pandas as pd
import numpy as np
import random

# Define a list of supplement names
supplement_names = [
    "Vitamin D3", "Probiotics", "Fish oil", "Tumeric",
    "Iron", "Vitamin B12", "GoldenRoot", "Cinnamon",
    "Algae Powder", "Soy Protein", "Vitamin K", "Vitamin E",
    "Zinc", "Berberine", "Omega 3 Oil", "Prebiotics",
    "Creatine", "Elderberry", "Electrolytes", "Pycnogenol", "Lutein+Zeaxanthin",
    "Hibiscus", "Taurine", "Quercetin", "PGX", "Vitamin C", "Apple Cider Vinegar",
    "Fenugreek", "Milk Thistle"
]

# Calculate 20% of the DataFrame's length
num_rows_to_update = int(len(rcd_df) * 0.2)

# Randomly select 20% of the DataFrame's indices
indices_to_update = np.random.choice(rcd_df.index, size=num_rows_to_update, replace=False)

# Update the selected rows
for idx in indices_to_update:
    # Randomly select a supplement name for each row
    selected_supplement = random.choice(supplement_names)
    current_value = rcd_df.at[idx, 'medications']
    # Check if the current value is NaN 
    if pd.isna(current_value):
        # If it's NaN, treat it as an empty string
        current_value = ""
    # Now safe to concatenate because current_value is guaranteed to be a string
    updated_value = current_value + "; " + selected_supplement if current_value else selected_supplement
    rcd_df.at[idx, 'medications'] = updated_value
```
